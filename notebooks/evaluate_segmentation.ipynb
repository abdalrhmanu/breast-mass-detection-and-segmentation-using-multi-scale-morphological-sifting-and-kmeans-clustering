```python
import os
import numpy as np
import cv2
```

```python
# Set the paths to the positive images and groundtruth folders
groundtruth_path = '../dataset/processed/groundtruth'

# Get a list of all the tif images in the positive images folder
groundtruth_dirs = [f for f in os.listdir(groundtruth_path) if f.endswith('.tif')]

groundtruth_dirs
```

```python
# Set the paths to the positive images and groundtruth folders
positive_segmented_path = '../dataset/output/segmentation'

# Get a list of all the tif images in the positive images folder
positive_dirs = [f for f in os.listdir(positive_segmented_path) if f.endswith('.tif')]

positive_dirs
```

```python
# Set the paths to the positive images and groundtruth folders
positive_kmeans_path = '../dataset/output/kmeans'

# Get a list of all the tif images in the positive images folder
positive_kmeans_dirs = [f for f in os.listdir(positive_kmeans_path) if f.endswith('.tif')]

positive_kmeans_dirs
```

```python
def calculate_iou(image1, image2):
    intersection = np.logical_and(image1, image2)
    union = np.logical_or(image1, image2)
    iou = np.sum(intersection) / np.sum(union)
    return iou

def calculate_sensitivity(image1, image2, threshold=0.5):
    iou = calculate_iou(image1, image2)
    sensitivity = iou >= threshold
    return sensitivity, iou
```

```python
ground_truth1 = os.path.join(groundtruth_path, '20587664_f4b2d377f43ba0bd_MG_R_ML_ANON.tif')
ground_truth2 = os.path.join(positive_segmented_path, '20587664_f4b2d377f43ba0bd_MG_R_ML_ANON_1.tif')

print(ground_truth1)
print(ground_truth2)

gt_img = cv2.imread(ground_truth1, cv2.IMREAD_GRAYSCALE)
pv_img = cv2.imread(ground_truth2, cv2.IMREAD_GRAYSCALE)

sensitivity, iou = calculate_sensitivity(gt_img, pv_img)
print(f"Sensitivity: {sensitivity}")
```

```python
results = []

for idx, f in enumerate(positive_dirs):
    im_name = f.split('.tif')[0][:-2]+'.tif'
    
    ground_truth1 = os.path.join(groundtruth_path, im_name)
    ground_truth2 = os.path.join(positive_segmented_path, f)
    
    gt_img = cv2.imread(ground_truth1, cv2.IMREAD_GRAYSCALE)
    pv_img = cv2.imread(ground_truth2, cv2.IMREAD_GRAYSCALE)

    sensitivity, iou = calculate_sensitivity(gt_img, pv_img)
    
#     print(f, im_name, sensitivity, iou)
    
    if iou > 0:
        print(f"Sensitivity: {sensitivity}, iou: {iou} on {f}")
        results.append(sensitivity) 
    
    
print(len(results))
```

```python
tp = results.count(True)
fn = results.count(False)

print(tp, fn)
```

```python
sensitivity = tp / (tp + fn)

sensitivity
```
